```python
users_file_path = "/Volumes/customer_revenue_analytics/cralf_wh/raw/cralf_users.csv"
```

```python
from pyspark.sql.types import *

users_schema = StructType([
    StructField("user_id", StringType()),
    StructField("signup_country", StringType()),
    StructField("signup_ts", StringType())
])

df_users = (
    spark.read.format("csv")
        .option("header","true")
        .schema(users_schema)
        .load(users_file_path)
)
```

```python
from pyspark.sql.functions import *

df_users_dedup = (
    df_users.dropDuplicates(['user_id'])
)
```

```python
df_clean = (
    df_users_dedup
        .withColumn('signup_ts', to_timestamp('signup_ts'))
)
```

```python
df_clean.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable(users_table)
```
