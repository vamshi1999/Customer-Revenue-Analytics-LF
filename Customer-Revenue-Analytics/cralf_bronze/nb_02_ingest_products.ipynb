```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
products_path = "/Volumes/customer_revenue_analytics/cralf_wh/raw/cralf_products.csv"
```

```python
products_schema = StructType([
    StructField("product_id", StringType()),
    StructField("category", StringType()),
    StructField("price", DoubleType())
])

products_df = (
    spark.read.format("csv")
        .option("header", "true")
        .schema(products_schema)
        .load(products_path)
)
```

```python
products_dedup = (
    products_df.dropDuplicates(['product_id'])
)
```

```python
products_clean = (
    products_dedup
        .filter(col("category").isNotNull())
        .filter(col("price") > 0)
)
```

```python
products_clean.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable(products_table)
```
